```python
df['desc_length'] = df['about_product'].apply(lambda x: len(x.split()))
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)   # remove special chars
    text = re.sub(r'\s+', ' ', text).strip() # remove extra spaces
    return text

df['About_prod_clean'] = df['about_product'].apply(clean_text)
```

```python
df = pd.get_dummies(df, columns=['category'], drop_first=True)
```

```python
from textblob import TextBlob

df['review_sentiment'] = df['review_content'].apply(
    lambda x: TextBlob(x).sentiment.polarity
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=300,       # limit features to avoid overfitting
    stop_words='english',   # remove common words
    ngram_range=(1,2)       # include unigrams + bigrams (stronger)
)
```

```python
tfidf_matrix = tfidf.fit_transform(df['About_prod_clean'])
```

```python
import pandas as pd

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out()
)
```

```python
df_final = pd.concat([df.reset_index(drop=True), tfidf_df], axis=1)
```

```python
df_final.drop(
    ['about_product', 'About_prod_clean', 'review_content', 'review_title'],
    axis=1,
    inplace=True
)
```

```python
df_final.drop(['rating_count'], axis=1, inplace=True)
```

```python
X = df_final.drop(['low_selling', 'product_name'], axis=1)
y = df_final['low_selling']
```
